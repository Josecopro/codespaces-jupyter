# Analisis de complejidad (Lineales y recursion simple)



## Ejemplo 1: Buscar objetivo x en una matriz n*m

**Descripcion exacta del problema**
Dada una matriz de enteros de tamano n*m y un valor objetivo x,
determinar si x aparece en la matriz usando una busqueda lineal por filas.


**Entradas y salidas (3 ejemplos)**

### Entrada 1
Matriz: $ [[3, 5, 2], [7, 1, 9], [4, 6, 8]] $
Objetivo x: 1
### Salida 1
Encontrado: True (x esta en la matriz)

### Entrada 2
Matriz: $ [[10, 11], [12, 13]] $
Objetivo x: 9
### Salida 2
Encontrado: False (x no esta en la matriz)

### Entrada 3
Matriz: $ [[2], [4], [6]] $
Objetivo x: 4
### Salida 3
Encontrado: True (x esta en la matriz)

**Complejidad esperada**
- Tiempo: O($n*m$) en el peor caso
- Espacio: O(1) extra

```python
def contains_target(matrix, x):
    for row in matrix:
        for value in row:
            if value == x:
                return True
    return False

# Ejemplos de uso
print(contains_target([[3, 5, 2], [7, 1, 9], [4, 6, 8]], 1))  # True
print(contains_target([[10, 11], [12, 13]], 9))               # False
print(contains_target([[2], [4], [6]], 4))                    # True
```

## Ejemplo 2: Bubble sort

**Descripcion exacta del problema**
Dado un arreglo de enteros, ordenar los elementos en orden ascendente usando ciclos y explicar la complejidad O($N$).


**Entradas y salidas (3 ejemplos)**

### Entrada 1
Arreglo: [5, 1, 4, 2, 8]
### Salida 1
Arreglo ordenado: [1, 2, 4, 5, 8]

### Entrada 2
Arreglo: [3, 0, -2, 7]
### Salida 2
Arreglo ordenado: [-2, 0, 3, 7]

### Entrada 3
Arreglo: [9, 9, 1, 2]
### Salida 3
Arreglo ordenado: [1, 2, 9, 9]

**Complejidad esperada**
- Tiempo: O(n^2) en el peor y promedio caso
- Espacio: O(1) extra

```python
def bubble_sort(nums):
    n = len(nums)
    # Recorre el arreglo varias veces, intercambiando vecinos
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
                swapped = True
        if not swapped:
            break
    return nums

# Ejemplos de uso
print(bubble_sort([5, 1, 4, 2, 8]))  # [1, 2, 4, 5, 8]
print(bubble_sort([3, 0, -2, 7]))    # [-2, 0, 3, 7]
print(bubble_sort([9, 9, 1, 2]))     # [1, 2, 9, 9]
```

## Ejemplo 3: Numero binario con bits alternados

**Descripcion exacta del problema (traduccion)**
Dado un entero positivo $n$, verificar si su representacion binaria tiene bits alternados,
es decir, si cada par de bits adyacentes es distinto. Se cumple $1 <= n <= 2^{31} - 1$.

**Entradas y salidas (3 ejemplos)**

### Entrada 1
n = 5
### Salida 1
True (5 en binario es 101)

### Entrada 2
n = 7
### Salida 2
False (7 en binario es 111)

### Entrada 3
n = 11
### Salida 3
False (11 en binario es 1011)

**Complejidad esperada**
- Tiempo: O($log_2(n)$)
- Espacio: O($log_2(n)$) por la cadena binaria

```python
def has_alternating_bits(n):
    bits = []
    while n > 0:
        bits.append(str(n % 2))
        n //= 2
    bits.reverse()
    for i in range(1, len(bits)):
        if bits[i] == bits[i - 1]:
            return False
    return True

# Ejemplos de uso
print(has_alternating_bits(5))   # True
print(has_alternating_bits(7))   # False
print(has_alternating_bits(11))  # False
```

# Fuerza bruta (Unicamente con ciclos)

## Ejercicio: Suma maxima de subarreglo (fuerza bruta)

**Descripcion exacta del problema**
Dado un arreglo de enteros, encontrar la suma maxima de cualquier subarreglo contiguo
usando fuerza bruta (dos ciclos anidados).

**Entradas y salidas (3 ejemplos)**

### Entrada 1
Arreglo: [1, -2, 3, 4, -1]
### Salida 1
Suma maxima: 7 (subarreglo [3, 4])

### Entrada 2
Arreglo: [-5, -1, -8]
### Salida 2
Suma maxima: -1 (subarreglo [-1])

### Entrada 3
Arreglo: [2, 3, -2, 5, -1]
### Salida 3
Suma maxima: 8 (subarreglo [2, 3, -2, 5])

**Complejidad esperada**
- Tiempo: O(n^2) en el peor caso
- Espacio: O(1) extra

```python
def max_subarray_sum_bruteforce(nums):
    max_sum = nums[0]
    n = len(nums)
    for i in range(n):
        current_sum = 0
        for j in range(i, n):
            current_sum += nums[j]
            if current_sum > max_sum:
                max_sum = current_sum
    return max_sum

# Ejemplos de uso
print(max_subarray_sum_bruteforce([1, -2, 3, 4, -1]))  # 7
print(max_subarray_sum_bruteforce([-5, -1, -8]))       # -1
print(max_subarray_sum_bruteforce([2, 3, -2, 5, -1]))  # 8
```

# Recursion simple (Decrecientes de recursion)

## Ejercicio: Different Ways to Add Parentheses (recursion)

**Descripcion exacta del problema**
Dada una expresion con numeros enteros y operadores $+ - *$,
calcular todos los posibles resultados al agregar parentesis en todas las formas posibles.


**Entradas y salidas (3 ejemplos)**

### Entrada 1
Expresion: 2-1-1
### Salida 1
Resultados: [0, 2]

### Entrada 2
Expresion: 2*3-4*5
### Salida 2
Resultados: [-34, -14, -10, -10, 10]

### Entrada 3
Expresion: 10+5
### Salida 3
Resultados: [15]

**Complejidad esperada**
- Tiempo: Exponencial en el numero de operadores (aprox. $O(C_n)$)
- Espacio: Exponencial por la cantidad de resultados, mas O(n) de recursion

```python
def diff_ways_to_compute(s, left=0, right=None):
    if right is None:
        right = len(s) - 1

    results = []

    for i in range(left, right + 1):
        if s[i] in "+-*":
            left_vals = diff_ways_to_compute(s, left, i - 1)
            right_vals = diff_ways_to_compute(s, i + 1, right)

            for a in left_vals:
                for b in right_vals:
                    if s[i] == "+":
                        results.append(a + b)
                    elif s[i] == "-":
                        results.append(a - b)
                    else:
                        results.append(a * b)

    # Si no hubo operadores, construimos el número manualmente
    if not results:
        num = 0
        for i in range(left, right + 1):
            num = num * 10 + int(s[i])
        return [num]

    return results


# Ejemplos de uso
print(diff_ways_to_compute("2-1-1"))     # [0, 2]
print(diff_ways_to_compute("2*3-4*5"))   # [-34, -14, -10, -10, 10]
print(diff_ways_to_compute("10+5"))      # [15]
```

## Ejercicio: Remove Nodes From Linked List (lista enlazada)

**Descripcion exacta del problema (traduccion)**
Dada la cabeza de una lista enlazada, elimina cada nodo que tenga un nodo con valor estrictamente mayor
a su derecha y devuelve la nueva cabeza.

**Entradas y salidas (3 ejemplos)**

### Entrada 1
Lista: [5, 2, 13, 3, 8]
### Salida 1
Lista: [13, 8] (5, 2 y 3 se eliminan porque hay valores mayores a la derecha)

### Entrada 2
Lista: [1, 1, 1]
### Salida 2
Lista: [1, 1, 1] (no hay valores mayores a la derecha)

### Entrada 3
Lista: [1, 2, 3, 4]
### Salida 3
Lista: [4] (cada nodo tiene un mayor a la derecha)

**Complejidad esperada**
- Tiempo: O(n)
- Espacio: O(1) extra

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverse_list(head):
    prev = None
    current = head
    while current:
        nxt = current.next
        current.next = prev
        prev = current
        current = nxt
    return prev

def remove_nodes(head):
    # Revertimos para recorrer de derecha a izquierda y mantener el maximo visto
    head = reverse_list(head)
    max_val = float("-inf")
    dummy = ListNode(0)
    tail = dummy
    current = head
    while current:
        if current.val >= max_val:
            max_val = current.val
            tail.next = current
            tail = current
        current = current.next
    tail.next = None
    return reverse_list(dummy.next)

# Funciones auxiliares para pruebas
def build_list(values):
    dummy = ListNode(0)
    tail = dummy
    for v in values:
        tail.next = ListNode(v)
        tail = tail.next
    return dummy.next

def list_to_array(head):
    out = []
    while head:
        out.append(head.val)
        head = head.next
    return out

# Ejemplos de uso
print(list_to_array(remove_nodes(build_list([5, 2, 13, 3, 8]))))  # [13, 8]
print(list_to_array(remove_nodes(build_list([1, 1, 1]))))          # [1, 1, 1]
print(list_to_array(remove_nodes(build_list([1, 2, 3, 4]))))       # [4]
```
